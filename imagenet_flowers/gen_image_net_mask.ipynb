```python
from xml.dom import minidom
import xml.etree.ElementTree as ET
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
import os
import glob
from tqdm import tqdm
```

```python
def gen_mask(file_path):
    mydoc = ET.parse(file_path)
    size = mydoc.getroot().find('size')
    W = int(size.find('width').text)
    H = int(size.find('height').text)
    mask = np.zeros([H,W])
    x,y = np.meshgrid(np.arange(0,W),np.arange(0,H))
    for bndbox_obj in mydoc.getroot().findall('object'):
        bndbox = bndbox_obj.find('bndbox')
        xmin = int(bndbox.find('xmin').text)
        xmax = int(bndbox.find('xmax').text)
        ymin = int(bndbox.find('ymin').text)
        ymax = int(bndbox.find('ymax').text) 
        #print(xmin,xmax,ymin,ymax)
        xc = (xmin+xmax)/2
        yc = (ymin+ymax)/2
        r2 = (2*(x-xc)/(xmax-xmin))**2 + (2*(y-yc)/(ymax-ymin))**2
        mask[r2<1] = 1.0
    return mask
```

```python
annotation_folder_paths = glob.glob('*/Annotation')
for annotation_folder_path in tqdm(annotation_folder_paths):
    flower_folder = os.path.split(annotation_folder_path)[0]
    mask_folder = os.path.join(flower_folder, 'mask')
    os.system('mkdir '+mask_folder)
    annotation_file_paths = glob.glob(annotation_folder_path+'/*.xml')
    for path in annotation_file_paths:
        mask = gen_mask(path)
        filepath = os.path.join(mask_folder, os.path.split(path)[1][:-4]+'.jpg')
        io.imsave(filepath, mask)
```
